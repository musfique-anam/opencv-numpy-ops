```python
#NumPy for Image Matrix Operations
#=================================

#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read Image
img=cv2.imread('images/sample.jpg')
```

```python
#Convert BGR -> RGB for display
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('results/gray.jpg', gray)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
#Bightness Increase
#Using NumPy array addition and clippling to keep values 0-255
bright = np.clip(gray + 20, 0, 255).astype(np.uint8)
cv2.imwrite('results/brightened.jpg', bright)

plt.imshow(bright, cmap='gray')
plt.title("Brightened Image (+20)")
plt.axis("off")
plt.show()
```

```python
#Brightness Increase - Method 2 (OpenCV convertScaleAbs)
bright_cv = cv2.convertScaleAbs(gray, alpha=1, beta=50)
cv2.imwrite('results/brightened_cv.jpg', bright_cv)

plt.imshow(bright_cv, cmap='gray')
plt.title("Brightened Image (cv2.convertScaleAbs)")
plt.axis("off")
plt.show()
```

```python
#Invert Image
inverted = 255-gray
cv2.imwrite('results/inverted.jpg', inverted)

plt.imshow(inverted, cmap='gray')
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```
